```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import initial_grid as ig
```

```python
carmap = ig.initial_grid(4)
```

```python
carmap.shape
```

```python
def plot(carmap, view=[]):
    if not len(view) ==0:
        dir_ = {'n': 2,
                'e': 0,
                's': 3,
                'w': 1}
        f = np.zeros_like(carmap[:,:,0])
        for l in view:
            f += carmap[:,:,dir_[l]]
        plt.imshow(f)
    else:
        plt.imshow(carmap.sum(axis=2))
    plt.colorbar()
```

```python
plot(carmap)
plt.title('Initial "Heatmap"')
```
